```python
def compress_image(X, K):
    train = X.copy().reshape(-1, X.shape[-1])
    centers, grouping = K_MEANS(train, K)
    used_colors = centers.reshape(K, train.shape[-1])
    for i in range(K):
        train[grouping == i] = used_colors[i]
    return train.reshape(X.shape)

def image_example(img_path, K):
    fig, (ax1, ax2) = plt.subplots(1,2, figsize=(10,10))
    X = imread(img_path)
    ax1.imshow(X)
    ax1.set_title('Standard picture')
    ax2.imshow(compress_image(X, K))
    ax2.set_title(f'Compressed with {K} colours')
```
